```python
import os
import matplotlib.pyplot as plt
from matplotlib import gridspec
from PIL import Image

def show_images_in_grid(folder_path, prefix, num_images=8):
    # Get a list of all files in the folder
    files = os.listdir(folder_path)
    
    # Use a list comprehension to filter files that contain the specified prefix
    gl_files = [file for file in files if prefix in file][:num_images]

    # Create a grid layout
    rows = 2
    cols = num_images // 2
    fig = plt.figure(figsize=(10, 5))
    gs = gridspec.GridSpec(rows, cols, wspace=0.1, hspace=0.1)

    for i, gl_file in enumerate(gl_files):
        img_path = os.path.join(folder_path, gl_file)
        img = Image.open(img_path)

        # Add subplot to the grid
        ax = plt.subplot(gs[i])
        ax.imshow(img)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(gl_file)

    plt.show()

print("Validation GLIDE:")
folder_path = "/home/enriconello/DeepFakeDetection/Thesis/3_image_generation/generated_images/validation"
show_images_in_grid(folder_path, "GL", num_images=10)

print("Test GLIDE:")
folder_path = "/home/enriconello/DeepFakeDetection/Thesis/3_image_generation/generated_images/test"
show_images_in_grid(folder_path, "GL", num_images=10)
```

```python
import os
import pandas as pd

# Read the CSV files
df = pd.read_csv("C:/Users/nello/Documents/vscode_projects/Thesis/3_image_generation/generated_images/csv/train_DL.csv")
df1 = pd.read_csv("C:/Users/nello/Documents/vscode_projects/Thesis/3_image_generation/generated_images/csv/train_SD.csv")
df2 = pd.read_csv("C:/Users/nello/Documents/vscode_projects/Thesis/3_image_generation/generated_images/csv/train_GL.csv")

# Merge the dataframes
merged_df = pd.concat([df, df1, df2], ignore_index=True)

# Get a list of valid IDs from the 'id' column
valid_ids = merged_df['id'].tolist()

# Path to the folder containing the images
folder_path = "C:/Users/nello/Documents/vscode_projects/Thesis/3_image_generation/generated_images/validation"

# Initialize counters
eliminated_count = 0
remaining_count = 0

# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    # Remove the file extension ".jpg"
    file_id_with_extension = os.path.splitext(filename)[0]
    
    # Extract the ID from the filename
    file_id = file_id_with_extension.split("_")[-1]  # Extract the last part after the last underscore
    
    # Check if the ID is not in the list of valid IDs
    if file_id not in valid_ids:
        
        # Construct the full path to the file
        file_path = os.path.join(folder_path, filename)
        
        # Delete the file
        os.remove(file_path)
        
        # Increment eliminated count
        eliminated_count += 1
    else:
        # Increment remaining count
        remaining_count += 1

# Print the counts
print("Number of eliminated images:", eliminated_count)
print("Number of remaining images:", remaining_count)
```
